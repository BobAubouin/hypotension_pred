```python
from scripts.test_haitem.feature_extraction import Extract_feature_from_segment
import pandas as pd
from pathlib import Path
from pyarrow.parquet import ParquetFile
import pyarrow as pa 

dataset_name = 'signal_dataset_haytem'
dataset_path = Path(f'data/datasets/{dataset_name}/cases/')
```

```python
pf = ParquetFile(dataset_path / 'case0001.parquet') 
first_ten_rows = next(pf.iter_batches(batch_size = 10)) 
data = pa.Table.from_batches([first_ten_rows]).to_pandas() 
```

```python

def test_number_sup(str,max_time):
    if 'time' in str or 'label' in str:
        return True
    if '_' in str:
        number = str.split('_')[1]
        if int(number) < max_time:
            return True
        else:
            return False
    else:
        return True

data = data[[col for col in data.columns if test_number_sup(col,150)]] # on restreint les données à 5 minutes
data.head()
```

```python
segment = data.iloc[0].copy()
segment = pd.DataFrame(segment).T
extrator = Extract_feature_from_segment(segment)
features = extrator.extract_all_features()

print(features)
```
